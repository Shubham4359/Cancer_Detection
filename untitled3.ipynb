```python
import numpy as np
import keras
#!pip install keras-self-attention
%config InlineBackend.figure_format = 'svg'
np.random.seed(0)
import tensorflow as tf
from tensorflow.keras import models
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Reshape, Add, Activation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D, Conv1D
from tensorflow.keras.layers import MaxPooling2D, MaxPooling1D, AveragePooling2D
from tensorflow.keras.layers import Flatten, GlobalAveragePooling2D, GlobalMaxPooling2D
from tensorflow.keras.layers import LeakyReLU, ReLU
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout, BatchNormalization
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad
from tensorflow.keras.layers import Lambda
import tensorflow.keras.backend as K
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

import random
random.seed(0)
```

```python
train_dir = '/content/data'
test_dir = '/content/testdata'
train_datagen = ImageDataGenerator(
                    validation_split=0.1,
                    rescale=1./255)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    class_mode='categorical',
                                                    target_size=(224, 224),
                                                    batch_size=32,
                                                    subset='training',
                                                    follow_links=True,
                                                    shuffle=True)

validation_datagen = train_datagen.flow_from_directory(train_dir,
                                                       class_mode='categorical',
                                                       target_size=(224, 224),
                                                       batch_size=32,
                                                       subset='validation',
                                                       follow_links=True,
                                                       shuffle=True)

test_datagen = test_datagen.flow_from_directory(test_dir,
                                                class_mode='categorical',
                                                target_size=(224, 224),
                                                batch_size=32,
                                                follow_links=True,
                                                shuffle=True)
```

```python
dataA = Input(shape = (224, 224, 3))

mid = keras.applications.ResNet50(weights = 'imagenet', input_shape= (224, 224, 3), include_top=False)(dataA)
flat = GlobalAveragePooling2D()(mid)

x = Dropout(0.3)(flat)
x = BatchNormalization()(x)
x = Dense(64, activation = 'relu')(x)
x = Dropout(0.35)(x)
out = Dense(3, activation = 'softmax')(x)

model = Model( inputs=dataA, outputs=out, name='breast_cancer' )
model.summary()

es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', mode='min',patience=3, verbose=1)

opt = Adam(learning_rate = 0.0001)
print("[INFO] compiling model...")  
model.compile(loss = "categorical_crossentropy", optimizer = opt, metrics=["accuracy"])
```

```python
history = model.fit(train_generator, epochs=5 ,batch_size = 128,validation_data=validation_datagen)
```

```python
import matplotlib.pyplot as plt
metrics = history.history
plt.plot(history.epoch, metrics['loss'], metrics['accuracy'])
plt.legend(['loss', 'categorical_accuracy'])
plt.show()
```

```python
test_labels = ['Benign', 'Malignant', 'Normal']
# Get predictions
pred = model.predict(test_datagen)

pred = np.argmax(pred, axis=-1)

labels = []

for i in test_datagen.filenames:
    if 'BIRAD1' in i:
        labels.append(0)
    elif 'Birad3' in i:
        labels.append(1)
    else:
        labels.append(2)

len(labels)
#return
# Original labels
from sklearn.metrics import accuracy_score
test_acc = accuracy_score(labels, pred)
print('Test accuracy:', test_acc)

from sklearn.metrics import classification_report
print(classification_report(labels, pred))
```

```python
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

cm  = confusion_matrix(labels, pred)
plt.figure()
plot_confusion_matrix(cm,figsize=(12,8), hide_ticks=True,cmap=plt.cm.Blues)
plt.xticks(range(3), test_labels, fontsize=16)
plt.yticks(range(3), test_labels, fontsize=16)
plt.show()
```

```python
def pad(img):

    nrows, ncols,d = img.shape

    # If padding is required...
    if nrows != ncols:

        # Take the longer side as the target shape.
        if ncols < nrows:
            target_shape = (nrows, nrows,d)
        elif nrows < ncols:
            target_shape = (ncols, ncols,d)

        # pad.
        padded_img = np.zeros(shape=target_shape)
        padded_img[:nrows, :ncols,:d] = img

    # If padding is not required...
    elif nrows == ncols:

        # Return original image.
        padded_img = img

    return padded_img
```
